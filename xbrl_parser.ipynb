```python
from lxml import etree

def xml_to_dict(xml_text):
    root = etree.fromstring(xml_text.encode())

    data = {}
    for elem in root.iter():
        tag = etree.QName(elem).localname
        
        if elem.text and elem.text.strip():
            data[tag] = elem.text.strip()
    
    return data
```

```python
def parse_outcome(data):
    return {
        "type": "outcome",
        "company": data.get("NameOfTheCompany"),
        "symbol": data.get("NSESymbol"),
        "event": data.get("FirstEventOfBoardMeeting"),
        "date": data.get("DateOfBoardMeetingConsideringDeclarationOfInterimOrRecommendationOfFinalDividend"),
        "dividend_type": data.get("FirstTypeOfDividend"),
        "rate": data.get("RateOfInterimDividendRecommendedPerEquityShare"),
        "record_date": data.get("RecordDateOfInterimDividend")
    }
```

```python
def parse_cirp(data):
    return {
        "type": "cirp",
        "company": data.get("NameOfTheCompany"),
        "symbol": data.get("NSESymbol"),
        "event": data.get("TypeOfEventOfCIRP"),
        "date": data.get("DateOfInitiationOfCIRP"),
        "authority": data.get("NameOfInsolvencyProfessional"),
        "details": data.get("BriefDetailsOfCIRP")
    }
```

```python
def format_for_finbert(parsed):
    t = parsed["type"]

    if t == "order_award":
        return f"{parsed['company']} received order worth {parsed['value']} from {parsed['entity']}."

    elif t == "outcome":
        return f"{parsed['company']} announced {parsed['event']} with dividend {parsed['rate']}."

    elif t == "cirp":
        return f"{parsed['company']} is undergoing insolvency proceedings under CIRP."

    return ""
```
