```python
import requests
from bs4 import BeautifulSoup
import time
from datetime import datetime

# Store seen links (to avoid duplicates)
seen_news = set()

URL = "https://www.moneycontrol.com/news/business/"

HEADERS = {
    "User-Agent": "Mozilla/5.0"
}

def fetch_news():
    try:
        res = requests.get(URL, headers=HEADERS, timeout=10)
        soup = BeautifulSoup(res.text, "html.parser")

        articles = soup.find_all("li", class_="clearfix")

        news_list = []

        for article in articles:
            title_tag = article.find("h2")
            link_tag = article.find("a")

            if title_tag and link_tag:
                title = title_tag.text.strip()
                link = link_tag["href"]

                if link not in seen_news:
                    seen_news.add(link)

                    news_list.append({
                        "title": title,
                        "url": link,
                        "time": datetime.now().strftime("%H:%M:%S")
                    })

        return news_list

    except Exception as e:
        print("Error:", e)
        return []


def run_realtime_scraper(interval=30):
    print("🚀 Starting Moneycontrol Real-Time Scraper...\n")

    while True:
        news_items = fetch_news()

        if news_items:
            print(f"\n📰 {len(news_items)} NEW ARTICLES FOUND:\n")

            for news in news_items:
                print(f"[{news['time']}] {news['title']}")
                print(news['url'])
                print("-" * 80)

        else:
            print("No new articles...")

        time.sleep(interval)


# Run scraper (every 30 seconds)
run_realtime_scraper(interval=30)
```
