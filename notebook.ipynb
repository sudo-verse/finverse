```python
company_to_ticker = {
    "Reliance Industries": "RELIANCE.NS",
    "Infosys": "INFY.NS",
    "TCS": "TCS.NS",
    "Wipro": "WIPRO.NS",
    "American Airlines": "AAL",
    "Meta": "META",
    "Bank of America": "BAC",
    "Pershing Square": "PSH"
}
```

```python
for article in articles[:10]:
    title = article["title"]

    # Sentiment
    sentiment_model = pipeline("sentiment-analysis")
    sentiment = sentiment_model(title)[0]["label"]

    # NER
    doc = nlp(title)
    companies = [ent.text for ent in doc.ents if ent.label_ == "ORG"]

    for company in companies:
        if company in company_to_ticker:
            ticker = company_to_ticker[company]

            stock = yf.Ticker(ticker)
            data = stock.history(period="1d")

            if not data.empty:
                price = data["Close"].iloc[-1]

                print("News:", title)
                print("Company:", company)
                print("Ticker:", ticker)
                print("Price:", price)
                print("Sentiment:", sentiment)
                print("-" * 50)
```

```python
def generate_signal(sentiment, score):
    if sentiment == "POSITIVE" and score > 0.7:
        return "BUY"
    elif sentiment == "NEGATIVE" and score > 0.7:
        return "SELL"
    else:
        return "HOLD"
```

```python
for article in articles[:10]:
    title = article["title"]

    sentiment = sentiment_model(title)[0]["label"]
    result = sentiment_model(title)[0]



    doc = nlp(title)
    companies = [ent.text for ent in doc.ents if ent.label_ == "ORG"]

    for company in companies:
        if company in company_to_ticker:
            ticker = company_to_ticker[company]

            stock = yf.Ticker(ticker)
            data = stock.history(period="1d")

            if not data.empty:
                price = data["Close"].iloc[-1]

                signal = generate_signal(result["label"], result["score"])

                print("News:", title)
                print("Company:", company)
                print("Price:", price)
                print("Sentiment:", sentiment)
                print("Signal:", signal)
                print("-" * 50)
```

```python
def risk_management(price):
    stop_loss = price * 0.98   # 2% loss
    target = price * 1.04      # 4% gain
    return stop_loss, target
```
